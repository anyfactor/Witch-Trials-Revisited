```python
import pandas as pd
```

### Getting the data

```python
df = pd.read_csv('https://raw.githubusercontent.com/JakeRuss/witch-trials/master/data/trials.csv')
df.head()
```

### Grouping by Country with count

```python
# grouping data
grouped_df = df.groupby(by="gadm.adm0").count()

# Keeping the essential data
grouped_df = grouped_df[["tried", "deaths"]]

# renaming column and resetting index
grouped_df = grouped_df.reset_index().rename({'gadm.adm0': 'country',}, axis=1)
grouped_df = grouped_df.set_index('country')
grouped_df
```

### Getting tried to death ratio and sorting by death

```python
grouped_df['death to tried ratio'] = grouped_df['deaths'] / grouped_df['tried']
grouped_df = grouped_df.sort_values('deaths', ascending=False)
grouped_df.to_csv("witch trials grouped.csv", index=False)
```
